#### iii. Time Series Segments

#### The Right and the Wrong Way to Do Cross-validation: 

The wrong way to perform cross validation is by including the samples on which the classifier (i.e. the k group) is to be run in the feature selection step.

The right way to perform cross validation is by making sure the samples on which the classifier (i.e. the k group) is to be run should be left out during the feature selection step. This ensures that the predictors are not biased and the prediction would be natural. 

```python
pairs={}
for w in np.arange(1,21):
    train_table_x=pd.DataFrame(data=np.arange(1,70), columns=['Instance'])
    t=np.arange(1,w+1)
    for l in t:
        for i in [0,1,5]:
            c=[]
            e=[]
            f=[]

            for j in range(0,len(train_df1), 480):
                if (j%480==0):
                    ch=train_df1.iloc[j+int(480/len(t)*(l-1)):int(j+480/len(t)*l), i]
                else:
                    ch=train_df1.iloc[j+int(480/len(t)*(l-1)):int(-1+j+480/len(t)*l), i]
                c.append(ch.mean())
                e.append(np.std(ch))
                f.append(ch.quantile(0.25))

            for j in range(0,len(train_df2), 480):
                ch=train_df2.iloc[j+int(480/len(t)*(l-1)):j+int(480/len(t)*l), i]
                c.append(ch.mean())
                e.append(np.std(ch))
                f.append(ch.quantile(0.25))

            train_table_x['mean_'+str(i+1)+'_'+str(l)]=c
            train_table_x['standard_deviation_'+str(i+1)+'_'+str(l)]=e
            train_table_x['1st_quartile_'+str(i+1)+'_'+str(l)]=f

    train_table_x.set_index('Instance', inplace=True)

    # bending = 1, other activities = 0
    train_table_x['y'] = 9*[1]+60*[0]
    lor = LogisticRegression(penalty= 'none', max_iter= 10000, random_state=0)
    lor_RFECV = RFECV(estimator=lor, cv=StratifiedKFold(5))
    lor_RFECV.fit(train_table_x.iloc[:, :-2], train_table_x.iloc[:, -1])
    score=lor_RFECV.score(train_table_x.iloc[:, :-2], train_table_x.iloc[:, -1])
    pairs[str(w)+":"+str(len(lor_RFECV.get_feature_names_out()))]=score

pairs
```

After using RFECV with StratifiedKFold(5), the best value of the pair
(l, p) comes out to be (10,68). 

Here the data didn't get to much fragmented as in case where l is 20, to loose its meaning. 
Also, the selection of 68 features out of 90 seems to be a good enough selection out of so many options.
Finally, the number of p selected that is the number of features, is 68, which is neither too low to cause bias leading to underfitting, nor too much to cause variance leading to overfitting. (Although it causes overfitting as we will see further due to class imbalance).
We cannot select the pair which has p greater than the number of samples, it would give no unique solution, so rejecting those.

```python
train_table_y=pd.DataFrame(data=np.arange(1,70), columns=['Instance'])
k=0
t=np.arange(1,11)
for l in t:
    for i in [0,1,5]:
        c=[]
        e=[]
        f=[]

        for j in range(0,len(train_df1), 480):
            if (j%480==0):
                ch=train_df1.iloc[j+int(480/len(t)*(l-1)):int(j+480/len(t)*l), i]
            else:
                ch=train_df1.iloc[j+int(480/len(t)*(l-1)):int(-1+j+480/len(t)*l), i]
            c.append(ch.mean())
            e.append(np.std(ch))
            f.append(ch.quantile(0.25))

        for j in range(0,len(train_df2), 480):
            ch=train_df2.iloc[j+int(480/len(t)*(l-1)):j+int(480/len(t)*l), i]
            c.append(ch.mean())
            e.append(np.std(ch))
            f.append(ch.quantile(0.25))

        train_table_y['mean_'+str(i+1)+'_'+str(l)]=c
        train_table_y['standard_deviation_'+str(i+1)+'_'+str(l)]=e
        train_table_y['1st_quartile_'+str(i+1)+'_'+str(l)]=f

train_table_y.set_index('Instance', inplace=True)

# bending = 1, other activities = 0
train_table_y['y'] = 9*[1]+60*[0]
lor1 = LogisticRegression(penalty= 'none', max_iter= 1000, random_state=0)
lor1_RFECV = RFECV(estimator=lor1, cv=StratifiedKFold(5))
lor1_RFECV.fit(train_table_y.iloc[:, :-2], train_table_y.iloc[:, -1])
score1=lor1_RFECV.score(train_table_y.iloc[:, :-2], train_table_y.iloc[:, -1])
print("\nScore: ", score1)
print("\nNumber of features selected: ", len(lor1_RFECV.get_feature_names_out()))
print("\nNames of the features selected: ", lor1_RFECV.get_feature_names_out())
print("\n", lor1_RFECV.support_)
print("\n", lor1_RFECV.ranking_)
```
