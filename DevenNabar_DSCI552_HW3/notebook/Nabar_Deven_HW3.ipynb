### (c) Feature Extraction

#### i. Research

The features usually used in Time Series Classification are minimum, maximum, mean, median, standard deviation, first quartile, and third quartile 

#### ii. Extraction

```python
folder = '../data/AReM/bending1'
files=glob.glob(os.path.join(folder+"/*."+"csv"))
files.sort(key=lambda x:[int(c) if c.isdigit() else c for c in re.split(r'(\d+)', x)])
#print(files)

li=[]
for i in files[0:]:
    bending1=pd.read_csv(i,index_col=False, skiprows=4)
    gh={}
    gh['Min1']=bending1.iloc[:,1:2].min()
    gh['Max1']=bending1.iloc[:,1:2].max()
    gh['Mean1']=bending1.iloc[:,1:2].mean()
    gh['Median1']=bending1.iloc[:,1:2].median()
    gh['standard deviation1']=bending1.iloc[:,1:2].std()
    gh['1st Quart 1']=bending1.iloc[:,1:2].quantile(0.25)
    gh['3rd Quart 1']=bending1.iloc[:,1:2].quantile(0.75)
    gh['Min2']=bending1.iloc[:,2:3].min()
    gh['Max2']=bending1.iloc[:,2:3].max()
    gh['Mean2']=bending1.iloc[:,2:3].mean()
    gh['Median2']=bending1.iloc[:,2:3].median()
    gh['standard deviation2']=bending1.iloc[:,2:3].std()
    gh['1st Quart 2']=bending1.iloc[:,2:3].quantile(0.25)
    gh['3rd Quart 2']=bending1.iloc[:,2:3].quantile(0.75)
    gh['Min3']=bending1.iloc[:,3:4].min()
    gh['Max3']=bending1.iloc[:,3:4].max()
    gh['Mean3']=bending1.iloc[:,3:4].mean()
    gh['Median3']=bending1.iloc[:,3:4].median()
    gh['standard deviation3']=bending1.iloc[:,3:4].std()
    gh['1st Quart 3']=bending1.iloc[:,3:4].quantile(0.25)
    gh['3rd Quart 3']=bending1.iloc[:,3:4].quantile(0.75)
    gh['Min4']=bending1.iloc[:,4:5].min()
    gh['Max4']=bending1.iloc[:,4:5].max()
    gh['Mean4']=bending1.iloc[:,4:5].mean()
    gh['Median4']=bending1.iloc[:,4:5].median()
    gh['standard deviation4']=bending1.iloc[:,4:5].std()
    gh['1st Quart 4']=bending1.iloc[:,4:5].quantile(0.25)
    gh['3rd Quart 4']=bending1.iloc[:,4:5].quantile(0.75)
    gh['Min5']=bending1.iloc[:,5:6].min()
    gh['Max5']=bending1.iloc[:,5:6].max()
    gh['Mean5']=bending1.iloc[:,5:6].mean()
    gh['Median5']=bending1.iloc[:,5:6].median()
    gh['standard deviation5']=bending1.iloc[:,5:6].std()
    gh['1st Quart 5']=bending1.iloc[:,5:6].quantile(0.25)
    gh['3rd Quart 5']=bending1.iloc[:,5:6].quantile(0.75)
    gh['Min6']=bending1.iloc[:,6:7].min()
    gh['Max6']=bending1.iloc[:,6:7].max()
    gh['Mean6']=bending1.iloc[:,6:7].mean()
    gh['Median6']=bending1.iloc[:,6:7].median()
    gh['standard deviation6']=bending1.iloc[:,6:7].std()
    gh['1st Quart 6']=bending1.iloc[:,6:7].quantile(0.25)
    gh['3rd Quart 6']=bending1.iloc[:,6:7].quantile(0.75)
    jk=pd.DataFrame.from_dict(gh)
    li.append(jk)
bending1=pd.concat(li,axis=0,ignore_index=True)
bending1 = bending1.apply(lambda x: pd.Series(x.dropna().values))
bending1.index=bending1.index + 1
bending1.index.name='Instance'
#bending1
```
